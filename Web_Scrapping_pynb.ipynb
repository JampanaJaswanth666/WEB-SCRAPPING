```python
import requests
from bs4 import BeautifulSoup
```

```python
# Identitiy the url
URL = "https://www.amrita.edu/"
```

```python
# Loading the webpage in memory using requests library

page = requests.get(URL)
```

```python
# check the status code of the page
page.status_code
```

```python
# Extracting the HTML code of the webpage

htmlcode = page.text
soup = BeautifulSoup(htmlcode)
```

```python
htmlcode
```

```python
content = soup.find('div', attrs=('class' , 'mobile-menu-content'))
print(content)
```

```python
text = content.text.strip()
print(text)
```

```python
content = soup.find_all('a', attrs=('class' , 'menu-link sub-menu-link'))
print(content)
```

```python
text = []
for i in content:
  text.append(i.text.strip())
print(text)
```

```python
Undergraduate = text[text.index('Undergraduate'):text.index('Medicine')-1]
Postgraduate = text[text.index('Postgraduate'):text.index('Integrated Degree')-1]
DegreePrograms = text[text.index('Degree Programs'):text.index('Curriculum')-1]
print(Undergraduate)
print(Postgraduate)
print(DegreePrograms)
```

```python
vd = Undergraduate.split('\n')
ve = Postgraduate.split('\n')
vf = DegreePrograms.split('\n')
```

```python
mainn = []
main1 = []
main2 = []
for i in vd:
  if i != '' and i != 'Undergraduate':
    mainn.append(i)
for i in ve:
  if i != '' and i != 'Postgraduate':
    main1.append(i)
for i in vf:
  if i != '' and i != 'Degree Programs':
    main2.append(i)
print(mainn)
print(main1)
print(main2)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Undergraduate':mainn, 'Postgraduate':main1, 'DegreePrograms':main2})
```
